####Get Info Box(Store in Python dictionary)



```python
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_best-selling_video_games")
```

```python
# convert to beautiful soup object
soup = bs(r.content)
```

```python
# Print out HTML
contents = soup.prettify
print(contents)
```

```python
info_box = soup.find(class_="wikitable plainrowheaders sortable")
```

```python
print(info_box)
```

```python
info_box.find_all('th')[8:]
```

```python
# Get an ordered list of game's rank 
def get_rank(row_data):
  rank =[int(th.get_text("", strip=True)) for th in info_box.find_all('th')[8:]]
  # Check for rowspans and append to list of ranks  
  for i,th in enumerate(info_box.find_all('th')[8:]):
    if th.has_attr("rowspan"):
      for j in range(int(th["rowspan"])):
        rank.append(int(th.get_text("",strip=True))) 
  rank.sort()
  return rank
```

```python
get_rank(info_box)
```

```python
def get_keys(row_data):
  tr = row_data.find('tr')
  return [th.get_text("",strip=True).replace('[a]', "") for th in tr.find_all('th')] 
```

```python
keys = get_keys(info_box)
keys
```

```python
def get_data(row_data):
  tbody = [td.get_text("", strip=True) for td in info_box.find("tbody").find_all('td')]
  
  # remove elements like [4][5] [b] [6]
  regex = re.compile(r"\[\w+]")
  game_data = [i.split('[')[0] for i in tbody if not regex.match(i)]
  # split game_data into list of each game data
  data_split = [game_data[i:i+6] for i in range(0,len(game_data),6)]
  
  rank = get_rank(row_data)
  # Add game rank to the data
  for i in range(len(data_split)):
    data_split[i].insert(0, rank[i])

  game_list = []
  keys = get_keys(row_data)
  for i in range(len(data_split)):
    game_list.append(dict(zip(keys,data_split[i])))
  
  return game_list
```

```python
game_list = get_data(info_box)
```

```python
game_list
```

### Clean data

```python
print([game.get('Sales','NAN') for game in game_list])
```

```python
# Convert sales to numbers
for game in game_list:
  game['Sales'] = int(game.get('Sales','NAN').replace(",","")) 
```
